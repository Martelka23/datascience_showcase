```python
import numpy as np 
import pandas as pd 
import os
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import missingno as msno

%matplotlib inline
```

```python
df = pd.read_csv('fatal-police-shootings-data.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df['month'] = pd.to_datetime(df['date']).dt.month
df['year'] = pd.to_datetime(df['date']).dt.year
```

```python
msno.matrix(df)
```

```python
df.isna().sum()/df.shape[0]*100
```

```python
# Since missing values are less than 5% for almost all the data. 
# Therefore, By dropping missing records we aren't loosing lot of information.
df.dropna(inplace=True)
```

```python
cardinality = {}
for key in df.columns:
    cardinality[key] = df[key].nunique()
cardinality
```

```python
# A - Asian, W - White, H - Hispanic, B - Black, O - Others, N - Native
races = df.race.unique()
races
```

```python
killing_by_race = pd.DataFrame(df['race'].value_counts())
killing_by_race = killing_by_race.reset_index()
fig = px.pie(killing_by_race, values='race', names='index')
fig.show()
```

```python
unarmed_shotout = pd.DataFrame(df[df['armed'] == 'unarmed']['race'].value_counts())
unarmed_shotout = unarmed_shotout.reset_index()
fig = px.pie(unarmed_shotout, values='race', names='index')
fig.show()
```

```python
unarmed_race_percent = {}

for race in df['race'].unique():
    total = df[df['race'] == race]['race'].count()
    unarmed = df[(df['race'] == race) & (df['armed'] == 'unarmed')]['race'].count()
    unarmed_race_percent[race] = unarmed / total * 100
    
unarmed_race_percent = pd.DataFrame(data=zip(unarmed_race_percent.keys(), unarmed_race_percent.values()), columns=['race', 'percent'])
fig = px.histogram(unarmed_race_percent, x='race', y='percent', color='race')
fig.show()
```

```python
shotout_by_states = df['state'].value_counts()
shotout_by_states = pd.DataFrame(shotout_by_states)
shotout_by_states = shotout_by_states.reset_index()
fig = px.histogram(shotout_by_states, x='index', y='state')
fig.show()
```

```python
kills_per_year = df[['year', 'race']]
kills_per_year['kills'] = 1
kills_per_year = kills_per_year.groupby(['year', 'race']).sum()
kills_per_year = kills_per_year.reset_index()
fig = px.bar(kills_per_year, x='year', y='kills', color='race', barmode='group')
fig.show()
```

```python
cities_top = df[['year', 'city']]
cities_top['kills'] = 1
cities_top = cities_top.groupby(['year', 'city']).sum()
cities_top = cities_top.reset_index()

cities_max = pd.DataFrame(columns=['year', 'city', 'kills'])
for year in range(2015, 2021):
    index = cities_top[cities_top['year'] == year]['kills'].idxmax()
    cities_max = cities_max.append(cities_top.loc[index], ignore_index=True)

fig = px.bar(cities_max, x='year', y='kills', color='city')
fig.show()
```
