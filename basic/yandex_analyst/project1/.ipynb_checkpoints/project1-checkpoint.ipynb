# Исследование надёжности заёмщиков

Заказчик — кредитный отдел банка. Нужно разобраться, влияет ли семейное положение и количество детей клиента на факт погашения кредита в срок. Входные данные от банка — статистика о платёжеспособности клиентов.

Результаты исследования будут учтены при построении модели **кредитного скоринга** — специальной системы, которая оценивает способность потенциального заёмщика вернуть кредит банку.

## Шаг 1. Начало работы

### Импортирование библиотек

```python
import numpy as np
import pandas as pd
from scipy import stats
from pymystem3 import Mystem
from matplotlib import pyplot as plt

%matplotlib inline
plt.style.use('fivethirtyeight')
```

### Открытие файла и получение общей информации

```python
data = pd.read_csv('/datasets/data.csv')

print(data.info())
data
```

**Вывод**

## Шаг 2. Предобработка данных

### Обработка букв разного регистра

<big>Применим метод lower() ко всем строкам в столбце education</big>

**Вывод**

### Обработка отрицательных значений

<big>Обработаем столбец days_employed. Ещё на первом шаге было замечено, что большая часть ячеек имеет отрицательные значения. Применим ко всем ячейкам модуль</big>

```python
data['days_employed'] = data['days_employed'].abs()
```

**Вывод**
<br><br>

### Обработка выбросов

```python
data = data.drop(data[data['children'].isin([-1, 20])].index)
data['children'].value_counts()
```

<big>Посмотрим описание столбца **возрастов** и заметим, что у нас присутствуют люди, которым 0 лет.</big>

```python
data['dob_years'].describe()
```

<big>Так как возраст тоже не особо важен для анализа, просто присвоим медиану в нулевой возраст</big>

```python
data.loc[data['dob_years'] == 0, 'dob_years'] = data['dob_years'].median()
```

<big>Посмотрим какие значения присутствуют в столбце **пола**</big>

```python
data['gender'].value_counts()
```

<big>Нашли непонятный xna, он такой один, поэтому просто удалим его</big>

```python
data = data[data['gender'] != 'XNA']
```

**Вывод**
<br><br>
<big>Мы избавились от всех выбросов. Их было очень много и они бы достаточно сильно могли повлиять на исследование</big>

### Обработка пропусков
<big>Посмотрим на процент пропущенных значений. Мы имеем 26% пропусков в столбце days_employed и 10% в total_income, это достаточно большая часть таблицы, поэтому эти пропуски лучше будет заполнить, а не просто выкинуть. В остальных столбцах пропусков не замечено.<br>
В days_employed 16% пропусков были добавлены нами вручную, когда мы чистили нашу таблицу от выбросов. Изначальные 10% у days_employed и total_income могли появиться из-за технических сбоев или в связи с тем, что некоторые люди решили не рассказывать о себе всю информацию.</big>

```python
data.isna().sum() * 100 / data.shape[0]
```

<big>Так как столбец со стажем не особо важен для будущего анализа, просто присвоим пустым ячейкам значение медианы</big>

```python
data['days_employed'] = data['days_employed'].fillna(data['days_employed'].median())
```

<big>Обработаем столбец total_income. Для начала посмотрим на описание значений. Видим, что с ними никаких аномалий не надблюдается.
</big>

```python
data['total_income'].describe()
```

```python
income_types_medians = data.groupby('income_type')['total_income'].median()
data.loc[data['total_income'].isna(), 'total_income'] = \
    data[data['total_income'].isna()]['income_type'].apply(lambda x: income_types_medians[x])
```

**Вывод**
<br><br>
<big>Все пропущенные значения были обработаны и пропущенные значения были заполнены вручную</big>

### Замена типов данных

```python
data['days_employed'] = data['days_employed'].astype('int')
data['total_income'] = data['total_income'].astype('int')
data['dob_years'] = data['dob_years'].astype('int')
data.info()
```

**Вывод**

<big>Поймали сразу двух зайцев: изменили типы на правильные и избавились от странных дробных частей. Используем метод astype(), так как это проще всего.</big>

### Обработка дубликатов

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates().reset_index(drop=True)
```

**Вывод**

<big>Удалим дубликаты самым простым способом. Самая частая причина появления дубликатов - неправильная работа программ, которые собирают данные.</big>

### Лемматизация

<big>Проведём лемматизацию и посмотрим, какие цели указывали люди. Напишем функцию, которая принимает на вход строку, далее лемматизирует её. После лемматизации получаем список строк, который снова объединяем в одну строку. Применяем эту функцию для каждого значения в столбце purpose.</big>

```python
m = Mystem()

def lem(string):
    lemmas = m.lemmatize(string.strip())
    return ''.join(lemmas[:-1])

data['purpose'] = data['purpose'].apply(lambda x: lem(x))
data['purpose'].value_counts()
```

**Вывод**

<big>Из данного списка можно выделить 4 основные категории целей взятия кредита: Недвижимость, автомобиль, свадьба и образование</big>

### Категоризация данных

<big>По выявленным категориям из 2.6 напишем функцию-классификатор и применим её к столбцу purpose</big>

```python
def purpose_classifier(purpose):
    answer = ''
    
    if 'автомобиль' in purpose:
        answer = 'Автомобиль'
    elif 'свадьба' in purpose:
        answer = 'Свадьба'
    elif 'недвижимость' in purpose or 'жилье' in purpose:
        answer = 'Недвижимость'
    elif 'образование' in purpose:
        answer = 'Образование'
        
    return answer

data['purpose'] = data['purpose'].apply(lambda x: purpose_classifier(x))
data['purpose'].value_counts()
```

```python
data['income_class'] = pd.qcut(data['total_income'], 4, labels=['Низкий', 'Средний', 'Выше среднего', 'Высокий'])
data['income_class'].value_counts()
```

## Шаг 3. Анализ данных

- <big>**Есть ли зависимость между наличием детей и возвратом кредита в срок?**</big>

<big>Построим график зависимости доли кредитов с задолженностью от общего количества детей в семье.</big>

```python
children = data[['children', 'debt']].copy()
children['count'] = 1
children = children.groupby(['children', 'debt'])['count'].sum()
children = children.reset_index()
children.loc[11] = [5, 1, 0]
debts = [(children.loc[i, 'count'], children.loc[i + 1, 'count']) for i in range(0, 11, 2)]

x = np.arange(0, 6)  # Количество детей
y = [y / (x + y) for x, y in debts]  # Список доли задолженностей
plt.plot(x, y)
plt.xlabel('Количество детей в семье')
plt.ylabel('Доля задолженности')
plt.show()
```

<big>На графике прослеживается небольшая зависимость, которая резко падает, когда мы рассматриваем количество детей 5. Это связано с тем, что таких людей в нашей выборке очень мало и так совпало, что ни у одного из них нет задолженности.
<br>Переменные количества детей и доли задолженностей являются количественными и распределены нормально, поэтому посчитаем коэффициент корреляции Пирсона. Как известно он очень чувствителен к выбросам, поэтому во время подсчёта мы исключим людей с 5 детьми.
<br>Нулевая гипотеза будет гласить, что зависимости между количеством детей и долей задолженности нет, а альтернативная гипотеза говорит о том, что зависимость есть.
</big>

```python
_, p = stats.shapiro(y[:-1])

if p > 0.05:
    print('Это нормальное распределение')
else:
    print('Распределение не является нормальным')

corr, p = stats.pearsonr(x[:-1], y[:-1])

print(f'Коэффициент корреляции Пирсона: {corr:.3f}')
print(f'Статистическая значимость: {p:.3f}')
```

**Вывод**

<big>Статистическая значимость оказалась больше 5%, поэтому мы отклоняем альтернативную гипотезу и делаем вывод о том, что скорее всего зависимости между количеством детей и долей задолженности нет. Но по крайней мере мы можем заметить, что люди без детей имеют меньший процент задолженности.</big>

- <big>**Есть ли зависимость между семейным положением и возвратом кредита в срок?**</big>

<big>Построим круговые диаграммы для каждого семейного положения.</big>

```python
family_status = data[['family_status', 'debt']].copy()
family_status['count'] = 1
family_status = family_status.groupby(['family_status', 'debt'])['count'].sum()
family_status = family_status.reset_index()

debts = [(family_status.loc[i, 'count'], family_status.loc[i + 1, 'count']) for i in range(0, 9, 2)]

fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(20, 20))

for i in range(5):
    ax[i].pie(debts[i], labels=['Нет задолженности', 'Есть задолженность'],
                 shadow=True, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'},
                 startangle=-50)
    ax[i].set_title(f'{family_status.loc[i * 2, "family_status"]}')

plt.tight_layout()
plt.show()
```

**Вывод**

<big>Видим, что различия в исследовании **есть**. Больше всего проблем с кредитами у неженатых (незамужних) людей процент неуплат составляет 9.8%. Также проблемы наблюдаются у тех, кто живет в гражданском браке 9.3%.
<br>Интересное наблюдение, что меньше всего проблем имеют люди, которые попали не в самую простую жизненную ситуацию. Люди в разводе и вдовы имеют соответственно 7.1% 6.6% неуплат.
<br>Середину занимают женатые (замужные) люди с 7.6%</big>

- <big>**Есть ли зависимость между уровнем дохода и возвратом кредита в срок?**</big>

<big>Построим круговые диаграммы для каждого уровня дохода, который мы выделили на этапе категоризации. Также не забываем про масску, которую мы сделали, когда заполняли пустые значения медианой. Отбросим такие значения, чтобы он не испортили нам анализ.</big>

```python
# Ключ сортировки дохода
def key(income):
    a = ['Низкий', 'Средний', 'Выше среднего', 'Высокий']
    return a.index(income)

income = data[['income_class', 'debt']].copy()
income['count'] = 1
income = income.groupby(['income_class', 'debt'])['count'].sum()
income = income.reset_index()

income['for_sort'] = income['income_class'].apply(lambda x: key(x))
income = income.sort_values(by=['for_sort', 'debt']).reset_index(drop=True)


debts = [(income.loc[i, 'count'], income.loc[i + 1, 'count']) for i in range(0, 8, 2)]

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))

for i in range(2):
    for j in range(2):
        ax[i][j].pie(debts[i * 2 + j], labels=['Нет задолженности', 'Есть задолженность'],
                    shadow=True, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'},
                    startangle=-60)
        ax[i][j].set_title(income.loc[i * 4 + 2 * j, 'income_class'])

plt.tight_layout()
plt.show()
```

<big>Видим, что процент задолженности немного растёт от низкого заработка к среднему, а потом резко падает.
<br>Проведём корреляционный анализ. У нас переменная заработка является качественной ранговой, а переменная доли - количественной. Поэтому используем коэффициент корреляции Спирмена. Выдвенем альтернативную гипотезу о том, что зависимость между заработком и долей задолженности есть, тогда нулевая гипотеза будет говорить об обратном.</big>

```python
x = np.arange(0, 4)  # Ранги заработка
y = [y / (x + y) for x, y in debts]  # Список доли задолженностей

corr, p = stats.spearmanr(x, y)
print(f'Коэффициент корреляции Спирмена: {corr:.3f}')
print(f'Статистическая значимость: {p:.3f}')
```

**Вывод**

<big>**Вывод**
<br>Статистическая значимость равна 60% значит мы не можем отвергнуть нулевую гипотезу. Получается, достаточной зависимости между уровнем дохода и долей задолженностей нет. Хоть мы и видим разницу в 1.7% между средним и высоким уровнями.</big>

- <big>**Как разные цели кредита влияют на его возврат в срок?**</big>

<big>Построим для каждой цели круговую диаграмму.</big>

```python
purposes = data[['purpose', 'debt']].copy()
purposes['count'] = 1
purposes = purposes.groupby(['purpose', 'debt'])['count'].sum()
purposes = purposes.reset_index()

debts = [(purposes.loc[i, 'count'], purposes.loc[i + 1, 'count']) for i in range(0, 7, 2)]

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

for i in range(2):
    for j in range(2):
        ax[i][j].pie(debts[i * 2 + j], labels=['Нет задолженности', 'Есть задолженность'],
                    shadow=True, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'},
                    startangle=-50)
        ax[i][j].set_title(f'{purposes.loc[i * 4 + 2 * j, "purpose"]}')
        
plt.tight_layout()
plt.show()
```

**Вывод**

<big>Видим, доля возврата кредита имеет небольшую зависимость от цели. Автомобиль и образование имеют просрочки в 9.3%. Недвижимость и свадьба чуть меньше по 7.3% и 7.9% соответственно.</big>
