## –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö

### –û–±—Ä–∞–±–æ—Ç–∫–∞ –≤—ã–±—Ä–æ—Å–æ–≤

<big>–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤—ã–±—Ä–æ—Å–æ–≤</big>

```python
# –§—É–Ω–∫—Ü–∏—è —Å—Ä–µ–∑–∞ —Ç–∞–±–ª–∏—Ü—ã –ø–æ –¥–∞–Ω–Ω—ã–º –∏–∑ –∑–∞–¥–∞–Ω–Ω–æ–π –∫–æ–ª–æ–Ω–∫–∏
def slice(data, column, range, title='', xlabel='', ylabel='', info=False, hist=False):
    if range[0] is None:
        range[0] = float('-inf')
    if range[1] is None:
        range[1] = float('+inf')
    
    new_data = data[
        (data[column] >= range[0]) & 
        (data[column] <= range[1]) |
        (data[column].isna())
    ]
    
    if info:
        print(f'Count of elements Before: {data.shape[0]} After: {new_data.shape[0]}')
        
    if hist:
        ax = new_data.hist(column=column, bins=20, edgecolor='black')[0][0]
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.set_title(title)
        plt.show()
    
    return new_data

# –§—É–Ω–∫—Ü–∏—è —É–¥–∞–ª–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤ —Å –ø–æ–º–æ—â—å—é –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω–æ–≥–æ —Ä–∞–∑–º–∞—Ö–∞
def emission_treatment(data, column, distance=1.5, title='', xlabel='', ylabel='', info=False, hist=False):
    lower4 = data[column].quantile(0.25)
    upper4 = data[column].quantile(0.75)
    inter_quartile = upper4 - lower4
    range = [lower4 - distance * inter_quartile, upper4 + distance * inter_quartile]
    
    return slice(data, column=column, range=range, title=title, xlabel=xlabel, ylabel=ylabel, info=info, hist=hist)
```

–û–∫, –≤—ã–≥–ª—è–¥–∏—Ç –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –ª—é–±–æ–ø—ã—Ç–Ω–æüëÄ

### –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤

<big>–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–ø–æ–ª–Ω—è–µ—Ç –ø—Ä–æ–ø—É—Å–∫–∏ —Å –ø–æ–º–æ—â—å—é –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏</big>

```python
def show_pearson(data, factor, target):
    corr, p = stats.pearsonr(data[factor], data[target])
        
    print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞: {corr:.3f}')
    print(f'–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å: {p:.3f}')
    
    return corr, p

def show_spearman(data, factor, target):
    corr, p = stats.spearmanr(data[factor], data[target])
        
    print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –°–ø–∏—Ä–º–µ–Ω–∞: {corr:.3f}')
    print(f'–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å: {p:.3f}')
    
    return corr, p

def show_lr_coef(data, train_column, target_column):
    x_train = np.array(data[train_column]).reshape(-1, 1)
    y_train = np.array(data[target_column]).reshape(-1, 1)
    
    LR = LinearRegression()
    LR.fit(x_train, y_train)
    
    print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏: {LR.coef_[0][0]:.3f}')
    
    return LR

def show_scatter_and_regression(data, train_column, target_column, LR, xlabel='', ylabel=''):
    fig, ax = plt.subplots()
    
    ax.scatter(data[train_column], data[target_column], alpha=0.1)
    
    x_train = np.array(data[train_column]).reshape(-1, 1)
    x_axis = np.array([x_train.min(), x_train.max()]).reshape(-1, 1)
    ax.plot(x_axis, LR.predict(x_axis), color='red', label='–õ–∏–Ω–∏—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏')
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    
    plt.show()

def regression_filling(data, train_column, target_column, xlabel='', ylabel='', inplace=True):
    tmp = data[~data[target_column].isna()].copy()

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
    show_pearson(tmp, train_column, target_column)

    # –ú–µ–Ω—è–µ–º —Ç–∏–ø Series –Ω–∞ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä
    x_target = np.array(data[data[target_column].isna()][train_column]).reshape(-1, 1)

    # –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
    LR = show_lr_coef(tmp, train_column, target_column)

    # –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
    if x_target.shape[0] != 0 and inplace:
        y_target = LR.predict(x_target).reshape(1, -1)[0]
        data.loc[data[target_column].isna(), target_column] = y_target

    # –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–ª–æ—â–∞–¥–µ–π
    show_scatter_and_regression(tmp, train_column, target_column, LR, xlabel, ylabel)
```

```python
for column in ['kitchen_area', 'living_area', 'ceiling_height']:
    regression_filling(data, 'total_area', column, all_cols['total_area'], all_cols[column])
```

–î–∞, –∑–æ–Ω–∞ —Ü–µ–Ω—Ç—Ä–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ!

```python
center_flats = peters_flats[peters_flats['city_centers_nearest_km'] <= 8].copy()

factors = [
    ['total_area', 'pearson'],
    ['ceiling_height', 'pearson'],
    ['rooms', 'spearman'],
    ['city_centers_nearest_km', 'spearman'],
    ['floor_type', 'no'],
    ['weekday_exposition', 'no'],
    ['month_exposition', 'no'],
    ['year_exposition', 'no']
]
prices = 'last_price'

for factor, coef_type in factors:
    # –ü–æ–¥—Å—á—ë—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
    if coef_type == 'pearson':
        corr, p = stats.pearsonr(center_flats[factor], center_flats[prices])
        
        print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞: {corr:.3f}')
        print(f'–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å: {p:.3f}')
        
    elif coef_type == 'spearman':
        corr, p = stats.spearmanr(center_flats[factor], center_flats[prices])
        
        print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –°–ø–∏—Ä–º–µ–Ω–∞: {corr:.3f}')
        print(f'–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å: {p:.3f}')
    
    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
    fig, ax = plt.subplots()
    
    if coef_type == 'pearson':
        LR = LinearRegression()
        LR.fit(np.array(center_flats[factor]).reshape(-1, 1), np.array(center_flats[prices]).reshape(-1, 1))
        print(f'–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏: {LR.coef_[0][0]:.3f}')
        
        
        x_axis = np.array([center_flats[factor].min(), center_flats[factor].max()]).reshape(-1, 1)
        ax.scatter(center_flats[factor], center_flats[prices], alpha=0.1)
        ax.plot(x_axis, LR.predict(x_axis), color='red', label='–õ–∏–Ω–∏—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏')
        
        ax.legend()
        
    else:
        treatments = []
        keys = sorted(center_flats[factor].unique())

        for key in keys:
            treatments.append(center_flats[center_flats[factor] == key][prices].to_list())
            
        ax.boxplot(treatments)
        rotation = 45 if center_flats[factor].dtypes == 'object' else 0
        ax.set_xticklabels(keys, rotation=rotation)
    
    ax.set_title(all_cols[factor])
    ax.set_ylabel(all_cols[prices])
    
    plt.show()
```

### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

<big>–í—ã–¥–µ–ª–∏–º 10 –≥–æ—Ä–æ–¥–æ–≤ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π</big>

```python
pd.set_option('max_columns', None)
nn_data = data[data['locality_name'].isin(top_10['locality_name'])].copy()
nn_data.head()
```

<big>–£–¥–∞–ª–∏–º –Ω–µ–Ω—É–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã</big>

```python
columns = [
    'first_day_exposition',
    'floors_total',
    'floor',
    'price_per_meter',
    'days_exposition',
    'weekday_exposition',
    'living_area_proportion',
    'kitchen_area_proportion'
]

nn_data = nn_data.drop(columns, axis=1)
```

<big>–†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –≤—Ö–æ–¥–Ω—ã–µ –∏ —Ü–µ–ª–µ–≤—ã–µ</big>

```python
y = nn_data['last_price'].copy()
X = nn_data.drop('last_price', axis=1).copy()
```

<big>Onehot –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ</big>

```python
def onehot_code(df, column, prefix='_'):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    
    return df
```

```python
columns = ['is_apartment', 'studio', 'open_plan', 'locality_name', 'floor_type', 'month_exposition']
for column in columns:
    X = onehot_code(X, column, column)
X.head()
```

<big>–ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö</big>

```python
scaler = StandardScaler()

num_cols = ['total_images', 'total_area',
       'rooms', 'ceiling_height', 'living_area',
       'kitchen_area', 'balcony',
       'airports_nearest', 'city_centers_nearest',
       'parks_around3000', 'parks_nearest', 'ponds_around3000',
       'ponds_nearest', 'year_exposition']

# X[num_cols] = scaler.fit_transform(X[num_cols])
X = scaler.fit_transform(X)
X
```

<big>–†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏</big>

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

x_train = torch.tensor(np.array(x_train)).type(torch.float32)
y_train = torch.tensor(np.array(y_train)).type(torch.float32)

x_test = torch.tensor(np.array(x_test)).type(torch.float32)
y_test = torch.tensor(np.array(y_test)).type(torch.float32)
```

### –°–æ–∑–¥–∞–Ω–∏–µ –Ω–µ–π—Ä–æ—Å–µ—Ç–∏

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        
        self.layer1 = nn.Linear(x_train.shape[1], 24)
        self.layer2 = nn.Linear(24, 24)
        self.out = nn.Linear(24, 1)
        
    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = F.relu(self.out(x))
        
        return x
    
network = NeuralNetwork()

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(network.parameters(), lr=0.3)
```

<big>–û–±—É—á–µ–Ω–∏–µ –Ω–µ–π—Ä–æ—Å–µ—Ç–∏</big>

```python
for epoch in range(1, 4):
    print('–≠–ø–æ—Ö–∞', epoch)
    for x, target in zip(x_train, y_train):
        optimizer.zero_grad()
        output = network(x)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
```

<big>–¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–µ–π—Ä–æ—Å–µ—Ç–∏</big>

```python
total_loss = 0

for x, target in zip(x_test, y_test):
    output = network(x)
    loss = criterion(output, target)
    total_loss += loss
```

<big>–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π</big>
