#### Конвертирование даты

```python
def convert_date(data, column):
    sr = data[column].copy()
    sr = pd.to_datetime(sr, format='%Y-%m-%d')
    
    return sr

users_df['reg_date'] = convert_date(users_df, 'reg_date')
users_df['churn_date'] = convert_date(users_df, 'churn_date')
calls_df['call_date'] = convert_date(calls_df, 'call_date')
messages_df['message_date'] = convert_date(messages_df, 'message_date')
internet_df['session_date'] = convert_date(internet_df, 'session_date')
```

#### Конвертирование целочисленных значений

```python
def convert_int(data, column):
    sr = data[column].copy()
    sr = sr.apply(np.ceil)
    sr = sr.astype('int32')
    
    return sr

calls_df['duration'] = convert_int(calls_df, 'duration')
internet_df['mb_used'] = convert_int(internet_df, 'mb_used')
```

## Анализ данных

<big>**Функция для отрисовки гистограммы и полного её анализа**</big>

```python
def display_hist_statistic(data, column, bins=25):
    # Подготовка данных
    sr = data[column].copy()
    
    # Построение гистограммы
    ax = sr.hist(bins=bins, figsize=(12, 7), edgecolor='black', zorder=1)
    
    # Статистики
    mean = sr.mean()
    variance = np.var(sr)
    sd = np.sqrt(variance)
    lower4 = sr.quantile(0.25)
    median = sr.median()
    upper4 = sr.quantile(0.75)
    iqr = upper4 - lower4
    
    stats = 'Статистики:\n' +\
        f'M(X): {int(mean):,}\n' +\
        f'D(X): {int(variance):,}\n' +\
        f'sd: {int(sd):,}\n' +\
        f'q(0.25): {int(lower4):,}\n' +\
        f'Me(X): {int(median):,}\n' +\
        f'q(0.75): {int(upper4):,}\n' +\
        f'IQR: {int(iqr):,}'
    
    # Границы и параметры
    ymax = 1.05 * ax.dataLim.intervaly.max()
    xmin = ax.dataLim.intervalx.min()
    xmax = ax.dataLim.intervalx.max()
    alpha = 0.3
    
    # Области отклонения
    ax.axvspan(max(mean - sd, xmin), min(mean + sd, xmax), 0.05,
               alpha=alpha, color='green', label='Интервал 1 sd', zorder=0)
    ax.axvspan(max(mean - 2 * sd, xmin), max(mean - sd, xmin), 0.05,
               alpha=alpha, color='yellow', label='Интервал 2 sd', zorder=0)
    ax.axvspan(min(mean + sd, xmax), min(mean + 2 * sd, xmax), 0.05,
               alpha=alpha, color='yellow', zorder=0)
    ax.axvspan(max(mean - 3 * sd, xmin), max(mean - 2 * sd, xmin), 0.05,
               alpha=alpha, color='orange', label='Интервал 3 sd', zorder=0)
    ax.axvspan(min(mean + 2 * sd, xmax), min(mean + 3 * sd, xmax), 0.05,
               alpha=alpha, color='orange', zorder=0)
    ax.axvspan(xmin, max(mean - 3 * sd, xmin), 0.05,
               alpha=alpha, color='red', label='Интервал >3 sd', zorder=0)
    ax.axvspan(min(mean + 3 * sd, xmax), xmax, 0.05,
               alpha=alpha, color='red', zorder=0)
    
    # Отображение основных статистик
    ax.vlines(lower4, 0, ymax, color='orange', label='Квартили')
    ax.vlines(mean, 0, ymax, color='#48b010', label='Среднее')
    ax.vlines(upper4, 0, ymax, color='orange')
    ax.vlines(median, 0, ymax, color='r', label='Медианное')
    ax.hlines(0, max(lower4 - 1.5 * iqr, xmin), min(upper4 + 1.5 * iqr, xmax), linestyles='dashed', 
              color='orange', lw=5, label='Размах боксплот')
    ax.scatter([max(lower4 - 1.5 * iqr, xmin), min(upper4 + 1.5 * iqr, xmax)], [0, 0], 
              s=150, edgecolor='black', color='orange', marker='X', zorder=2)
    
    ax.scatter([], [], color='white', label=stats)
    
    # Настройка и вывод
    ax.set_title(coll_names[column])
    ax.set_ylabel('Кол-во пользователей')
    ax.set_ylim(-0.05 * ymax, ymax)
    ax.legend()
    
    plt.show()
```

## Проверка гипотез

### Подсчёт выручки

<big>**Разделение датафреймов**</big>

```python
smart_users_df = users_df[users_df['tariff'] == 'smart'].copy()
ultra_users_df = users_df[users_df['tariff'] == 'ultra'].copy()
```

<big>**Класс тарифа для подсчёта выручки**</big>

```python
class Tariff:
    # Конструктор
    def __init__(self, price, minutes, messages, internet, minute_price, message_price, internet_price):
        self.price = price
        self.minutes = minutes
        self.messages = messages
        self.internet = internet
        self.minute_price = minute_price
        self.message_price = message_price
        self.internet_price = internet_price
        
    # Сложение словарей
    def dict_sum(self, *args):
        keys, values = [], []
        answer = Counter()

        for d in args:
            keys.extend(d.keys())
            values.extend(d.values())

        for key, value in zip(keys, values):
            answer[key] += value

        return answer

    # Подсчёт переплат
    def overprice_counter(self, data, column, limit, overprice):
        data['spent'] = data[column].apply(lambda x, a=limit, b=overprice: max(x - a, 0) * b)
        
        return {key: value for key, value in zip(data['month'].values, data[column].values)}
    
    # Затраты для определённого абонента
    def expenses_for_person(self, calls_grouped, messages_grouped, internet_grouped):
        overprice_for_minutes = self.overprice_counter(calls_grouped, 'duration_per_month', self.minutes, self.minute_price)
        overprice_for_messages = self.overprice_counter(messages_grouped, 'count_per_month', self.messages, self.message_price)
        overprice_for_internet = self.overprice_counter(internet_grouped, 'gb_per_month', self.internet, self.internet_price)
        
        overprice_dict = self.dict_sum(overprice_for_minutes, overprice_for_messages, overprice_for_internet)
        expenses_dict = self.dict_sum(overprice_dict, dict.fromkeys(overprice_dict.keys(), self.price))
        
        return expenses_dict
    
    # Затраты для всех абонентов
    def expenses_for_dataframe(self, users_df, calls_grouped, messages_grouped, internet_grouped):
        expenses_df = pd.DataFrame(columns=['user_id', 'city', 'month', 'expenses'])
        
        for i, row in users_df.iterrows():
            user_id = row['user_id']
            city = row['city']
            
            user_dfs = map(lambda df: df[df['user_id'] == user_id], [calls_grouped, messages_grouped, internet_grouped])
            expenses_dict = self.expenses_for_person(*user_dfs)
            
            data = \
                zip([user_id] * len(expenses_dict), [city] * len(expenses_dict), expenses_dict.keys(), expenses_dict.values())
            new_df = pd.DataFrame(data=data, columns=['user_id', 'city', 'month', 'expenses'])
            expenses_df = pd.concat([expenses_df, new_df])
            
        expenses_df['expenses'] = expenses_df['expenses'].astype('int')
            
        return expenses_df
```

<big>**Объекты класса**</big>

```python
tariff_smart = Tariff(price=550, minutes=500, messages=50, internet=15, 
                      minute_price=3, message_price=3, internet_price=200)
tariff_ultra = Tariff(price=1950, minutes=3000, messages=1000, internet=30, 
                      minute_price=1, message_price=1, internet_price=150)
```

<big>**Подсчёт выручки**</big>

```python
smart_proceeds_df = \
  tariff_smart.expenses_for_dataframe(smart_users_df, calls_grouped, messages_grouped, internet_grouped).reset_index(drop=True)
smart_proceeds_df['tariff'] = 'Смарт'

ultra_proceeds_df = \
  tariff_ultra.expenses_for_dataframe(ultra_users_df, calls_grouped, messages_grouped, internet_grouped).reset_index(drop=True)
ultra_proceeds_df['tariff'] = 'Ультра'

total_proceeds_df = pd.concat([smart_proceeds_df, ultra_proceeds_df]).sort_values(by='user_id').reset_index(drop=True)
my_display(total_proceeds_df.head(10))
```

<big>Распределение тарифа Ультра нестандартное.</big>

<big>**Функция для проверки гипотез**</big>

```python
def check_hypothesis(first, second, first_name, second_name):
    
    result = st.ttest_ind(first, second, equal_var=False)

    print(formatter.bold + 'p-значение:' + formatter.reset, result.pvalue, '=>', end=' ')
    if result.pvalue < 0.03:
        print('нулевая гипотеза отклонена')
    else:
        print('нулевая гипотеза не может быть отклонена')

    print(formatter.bold + 'Среднее:' + formatter.reset)
    print(formatter.blue + f'{first_name}:' + formatter.reset, f'{int(first.mean()):,}')
    print(formatter.red + f'{second_name}:' + formatter.reset, f'{int(second.mean()):,}')
    print(formatter.bold + 'Общий заработок:' + formatter.reset)
    print(formatter.blue + f'{first_name}:' + formatter.reset, f'{int(first.sum()):,}')
    print(formatter.red + f'{second_name}:' + formatter.reset, f'{int(second.sum()):,}')
    
    ax = first.hist(bins=20, figsize=(12, 7), alpha=0.5, label=first_name)
    
    ymax = 1.05 * ax.dataLim.intervaly.max()
    ax.hist(second, bins=20, alpha=0.5, label=second_name)
    ax.vlines(first.mean(), 0, ymax, color='b', label=f'Среднее для {first_name}')
    ax.vlines(second.mean(), 0, ymax, color='r', label=f'Среднее для {second_name}')

    ax.set_xlabel('Затраты в месяц')
    ax.set_ylabel('Кол-во абонентов')
    ax.legend()

    plt.show()
```

<big>**Сравниваем выручки по тарифам в месяц**</big>

<big>
    <ul>
        <li><b>Нулевая гипотеза:</b> выручки по тарифам в месяц равны</li>
        <li><b>Альтернативная гипотеза:</b> выручки по тарифам в месяц не равны</li>
    </ul>
</big>

```python
check_hypothesis(smart_proceeds_df['expenses'], ultra_proceeds_df['expenses'], 'Смарт', 'Ультра')
```

<big>**Вывод:**</big>

<big>Исходя из статистического теста мы отклоняем нулевую гипотезу. Делаем вывод о том, что выручка по тарифу Ультра в месяц не равна выручке по Смарт. Зато общая выручка больше по тарифу Смарт. Скорее всего это связанно с тем, что тарифу Ультра не хватает рекламы. 
<br>Также, из графика видно, что средняя выручка по тарифу Ультра превышает выручку по тарифу Смарт.</big>

### Выручка по городам

    <ul>
        <li><b>Нулевая гипотеза:</b> выручки по городам в месяц равны</li>
        <li><b>Альтернативная гипотеза:</b> выручки по городам в месяц не равны</li>
    </ul>
</big>

```python
moscow_proceeds_smart = smart_proceeds_df[smart_proceeds_df['city'] == 'Москва']['expenses'].copy()
not_moscow_proceeds_smart = smart_proceeds_df[smart_proceeds_df['city'] != 'Москва']['expenses'].copy()

moscow_proceeds_ultra = ultra_proceeds_df[ultra_proceeds_df['city'] == 'Москва']['expenses'].copy()
not_moscow_proceeds_ultra = ultra_proceeds_df[ultra_proceeds_df['city'] != 'Москва']['expenses'].copy()

print(formatter.bold + 'Тариф Смарт:' + formatter.reset)
check_hypothesis(not_moscow_proceeds_smart, moscow_proceeds_smart, 'Другие города', 'Москва')
print(formatter.bold + 'Тариф Ультра:' + formatter.reset)
check_hypothesis(not_moscow_proceeds_ultra, moscow_proceeds_ultra, 'Другие города', 'Москва')
```
