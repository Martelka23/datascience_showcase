# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
# YOUR CODE
# 1 Обработаем данные
x_train = data[feature_cols].copy()
y_train = data[target_col].copy()

x_test = data_test[feature_cols].copy()
# y_test = data_test[target_col].copy()

# Нормировка
scaller = StandardScaler()
x_train[num_cols] = scaller.fit_transform(x_train[num_cols])
x_test[num_cols] = scaller.transform(x_test[num_cols])

# one-hot encoding
dummy_features = pd.get_dummies(x_train[cat_cols])
x_train = pd.concat([x_train[num_cols], dummy_features], axis=1)

dummy_features = pd.get_dummies(x_test[cat_cols])
x_test = pd.concat([x_test[num_cols], dummy_features], axis=1)
```

```python
searchCV = LogisticRegressionCV(
    Cs=[100, 10, 1, 0.1, 0.01, 0.001],
    penalty='l2',
    scoring='roc_auc',
    cv=5,
    refit=True,
    max_iter=1000
)
searchCV.fit(x_train, y_train)
print('Max auc_roc:', searchCV.scores_[1].max())
```

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с дефолтными параметрами даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметроа количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию. 

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

```python
import catboost
```

```python
# YOUR CODE
x_train, x_validation, y_train, y_validation = train_test_split(
    data[feature_cols], data[target_col], train_size=0.8
)
```

```python
# Использовал тут самодельную кроссвалидацию, поэтому модель построена на цикле
# Сейчас оставил самые оптимальные настройки
lrs = np.array([0.05])
nes = [500]

max_auc = 0

for lr in lrs:
    for ne in nes:
        boosting_model = catboost.CatBoostClassifier(
            n_estimators=ne, learning_rate=lr, cat_features=cat_cols, logging_level='Silent')
        boosting_model.fit(x_train, y_train)

        y_train_pred = boosting_model.predict_proba(x_train)[:, 1]
        y_validation_pred = boosting_model.predict_proba(x_validation)[:, 1]
        
        train_auc = roc_auc_score(y_train, y_train_pred)
        validation_auc = roc_auc_score(y_validation, y_validation_pred)
        
        if max_auc < validation_auc:
            max_auc = validation_auc
            best_lr = lr
            best_ne = ne
            print(max_auc, best_lr, best_ne)
```

```python
validation_auc
```

<big>Лучшее качество 0.85622 с параметрами: n_estimators=500, learning_rate=0.05</big>
